```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# importing data

```python
df = pd.read_csv("imdb.csv", encoding = 'latin-1')
```

```python
df
```

```python
df.describe()
```

# removing unwanted columns

```python
df = df.drop(columns=['cast', 'crew', 'tagline','homepage','belongs_to_collection', 'poster_path'])
```

```python
columns = df.columns.tolist()
print(columns)
```

```python
null_counts = df.isnull().sum()
print(null_counts)
```

# filling the mean value in runtime column

```python
df['runtime'].fillna(107.85, inplace=True)
```

```python
print(df)
```

# filing the missing values in the columns

```python
df['genres'].fillna('unknown_genres', inplace=True)
df['overview'].fillna('unknown_overview ', inplace=True)
df['production_companies'].fillna('unknown_production_companies  ', inplace=True)
df['production_countries'].fillna('unknown_production_countries ', inplace=True)
df['spoken_languages'].fillna('unknown_spoken_languages', inplace=True)
df['Keywords'].fillna('unknown_Keywords', inplace=True)
```

```python
null_counts = df.isnull().sum()
print(null_counts)
```

```python
print(df.dtypes)
```

# converting date and time format

```python
df['release_date'] = pd.to_datetime(df['release_date'], format='%m/%d/%y', errors='coerce')
df['release_date'] 
```

```python
print(df.dtypes)
```

```python
df
```

# Filtering Movies Released After 2020

```python
filtered_df = df[df['release_date'] > "2020-12-31"]
```

#  Percentage of Movies in Each Status

```python
status_counts = df['status'].value_counts(normalize=True) * 100
print(status_counts)
```

# Bar Chart: Movie Count by Language.

```python
movies_by_language = df['original_language'].value_counts()
plt.figure(figsize=(10, 6))
bars = movies_by_language.plot(kind='bar', color='red')
for bar in bars.patches:
    bars.annotate(f'{int(bar.get_height())}', (bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                  ha='center', va='bottom')
plt.title('Movie Count by Language')
plt.xlabel('Language')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

# Pie Chart: Distribution of Movie Status

```python
movies_by_status = df['status'].value_counts()

# Create pie chart
plt.figure(figsize=(6,6))
plt.pie(movies_by_status, labels=movies_by_status.index, autopct='%1.1f%%', startangle=140, colors=['lightgreen', 'red'])
plt.title('Distribution of Movie Status')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

# Histogram: Budget Distribution

```python
plt.figure(figsize=(10, 6))
plt.hist(df['budget'], bins=5, color='yellow', edgecolor='red')
plt.title('Budget Distribution')
plt.xlabel('Budget (in millions)')
plt.ylabel('Frequency(count)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.show()
```

# HeatMap: correlation heatmap of movie feature

```python
import seaborn as sns
correlation_matrix = df[['budget', 'popularity', 'runtime', 'revenue']].corr()

# Create heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Correlation Heatmap of Movie Features')
plt.show()
```

# Scatter Plot: Budget vs. Revenue

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['budget'], df['revenue'], color='red', alpha=0.6, edgecolors='w', linewidth=0.8)
plt.title('Budget vs. Revenue')
plt.xlabel('Budget (in dollars)')
plt.ylabel('Revenue (in dollars)')
plt.grid(True)
plt.show()
```
